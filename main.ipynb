```python
# Detecting outliers through Interquartile Range (IQR)
def detect_outliers(df):
    outliers_count = {}
    for column in df.columns:
        data = df[column]
        data = sorted(data)
            
        q1 = np.percentile(data, 25)
        q3 = np.percentile(data, 75)
            
        IQR = q3 - q1
        lower_bound = q1 - (1.5 * IQR)
        upper_bound = q3 + (1.5 * IQR)
            
        num_outliers = len([i for i in data if i < lower_bound or i > upper_bound])
        outliers_count[column] = num_outliers
    return outliers_count
```

```python
# Displaying the count of outliers for each column
outliers_count = detect_outliers(df_week)
for column, num_outliers in outliers_count.items():
    print(f"Outliers in column '{column}': {num_outliers}")
```

```python
def plot_histograms(df, columns, title, colors=None, ylabel="Frequency"):
    """
    Plots histograms with KDE for given columns in a DataFrame.

    Parameters:
    df (pd.DataFrame): DataFrame containing the data.
    columns (list): List of column names to plot.
    colors (list, optional): List of colors for each histogram. Defaults to None.
    title (str, optional): Title for the entire plot. Defaults to "Histogram Comparison".
    xlabel (str, optional): Label for the x-axis. Defaults to "Total Kilometers".
    ylabel (str, optional): Label for the y-axis. Defaults to "Frequency".
    """

    num_plots = len(columns)
    fig, axes = plt.subplots(1, num_plots, figsize=(6 * num_plots, 6))

    # Default colors if none provided
    if colors is None:
        colors = ['blue', 'green', 'red'][:num_plots]

    for i, col in enumerate(columns):
        sns.histplot(df[col], kde=True, bins=30, ax=axes[i], color=colors[i])
        axes[i].set_title(f'Distribution of {col}')
        axes[i].set_ylabel(ylabel)

    # Set a common title
    fig.suptitle(title, fontsize=14, fontweight='bold')

    # Adjust layout to prevent label overlap
    fig.subplots_adjust(hspace=0.3, wspace=0.3)
    plt.tight_layout(rect=[0, 0, 1, 0.96])

    plt.show()
```

```python
def create_pie_chart(suffix='', ax=None):
    """
    Plots pie charts for the intensity training zones.

    Parameters:
    suffix: Identify different weeks of the dataset columns.
    """
    
    # Define the running zones
    columns = {
        'Zone Z3-Z4-Z5-T1-T2': f'total km Z3-Z4-Z5-T1-T2{suffix}',
        'Zone Z3-4': f'total km Z3-4{suffix}',
        'Zone Z5-T1-T2': f'total km Z5-T1-T2{suffix}'
    }

    # Filter out missing columns
    intensity_zones = {zone: df_week[col].sum() for zone, col in columns.items() if col in df_week}

    # If no valid data, return
    if not intensity_zones:
        print(f"Warning: No valid data for suffix '{suffix}'")
        return

    # Create a pie chart
    ax.pie(
        intensity_zones.values(), labels=intensity_zones.keys(),
        autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99']
    )
    ax.set_title(f'Training Intensities ({suffix})')

# Create subplots for better visualization
fig, axes = plt.subplots(1, 3, figsize=(12, 4))  

# Generate pie charts in subplots
for i, suffix in enumerate(['', '.1', '.2']):
    create_pie_chart(suffix, axes[i])

plt.tight_layout()
plt.show()
```

<div style='line-height:1.8;'>
<p style='text-align: justify;'>
To explore the relationships between training intensity, success rates, and recovery times across different athletes, we first aggregate training data over three weeks. This ensures a more stable representation of each athlete’s overall training patterns rather than analyzing individual weekly fluctuations.
</p>
<p style='text-align: justify;'>
To achieve this, we compute the three-week average for key training intensity metrics (e.g., total kms, exertion levels), success rates, and recovery scores. Additionally, we aggregate these features at the athlete level to study long-term patterns rather than isolated weeks. 
</p>
<p style='text-align: justify;'>
This transformation allows us to identify trends such as whether <strong>higher training intensity correlates with perceived training success or leads to lower recovery scores</strong>, helping us understand how training load affects performance and well-being.
</p>
</div>

```python
# Aggregate the dataframe
aggregated_df = df_week.copy()

# Columns to be aggregated from the dataframe
columns_to_aggregate = ['nr. sessions', 'nr. rest days', 'total kms', 'max km one day',
                       'total km Z3-Z4-Z5-T1-T2',
                       'nr. tough sessions (effort in Z5, T1 or T2)',
                       'nr. days with interval session', 'total km Z3-4',
                       'max km Z3-4 one day', 'total km Z5-T1-T2', 'max km Z5-T1-T2 one day',
                       'total hours alternative training', 'nr. strength trainings',
                       'avg exertion', 'min exertion', 'max exertion', 'avg training success',
                       'min training success', 'max training success', 'avg recovery',
                       'min recovery', 'max recovery']

for column in columns_to_aggregate:
    aggregated_df[column + '_avg'] = (aggregated_df[column] + aggregated_df[column + '.1'] + aggregated_df[column + '.2']) / 3

# Drop the original columns that ends with '.1' and '.2'
columns_to_drop = [col for col in aggregated_df.columns if col.endswith(('.1', '.2'))]
aggregated_df.drop(columns=columns_to_drop, inplace=True)
aggregated_df.drop(columns=columns_to_aggregate, inplace=True)

# Display the aggregated dataframe
aggregated_df
```

```python
# Calculate the total average success, recovery, and exertion for each athlete according with their ID
athlete_agg = aggregated_df.groupby('Athlete ID').agg({
    'avg training success_avg': 'sum',
    'avg recovery_avg': 'sum',
    'avg exertion_avg': 'sum'
}).reset_index()

# Display the final aggregated dataframe
athlete_agg
```

### **5.3.1 Grouped Bar Plots**

<div style='line-height:1.8;'>
<p style='text-align: justify;'>
Grouped bar plots allow for <strong>direct comparison of tough sessions and interval sessions across multiple time points (three consecutive weeks)</strong>. Since the dataset involves <strong>categorical (discrete) data</strong>, such as the <strong>number of tough sessions and days with interval sessions</strong>, bar plots are well-suited for showing <strong>how often each value occurs</strong>.
</p>
</div>

```python
# Function to create grouped bar chart with proper alignment
def create_grouped_bar_chart(ax, data_columns, title, xlabel, ylabel):
    # Count frequencies for each column
    value_counts_dict = {}
    all_unique_values = set()
    
    for col in data_columns:
        # Get value counts and convert to dictionary
        value_counts = df_week[col].value_counts().sort_index()
        value_counts_dict[col] = value_counts
        all_unique_values.update(value_counts.index)
    
    # Sort all unique values
    all_unique_values = sorted(all_unique_values)
    
    # Number of groups and bars
    n_values = len(all_unique_values)
    n_columns = len(data_columns)
    
    # Set width of bars
    bar_width = 0.8 / n_columns
    
    # Colors for the bars
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c']
    
    # Create bars with proper positions
    for i, col in enumerate(data_columns):
        # Create a frequency array that includes zeros for missing values
        frequencies = [value_counts_dict[col].get(val, 0) for val in all_unique_values]
        
        # Position for this group of bars
        positions = [x + i * bar_width for x in range(n_values)]
        
        # Create bars
        ax.bar(positions, frequencies, width=bar_width, alpha=0.7, 
               label=col, color=colors[i], edgecolor='black')
    
    # Set axis labels and title
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    
    # Set x-ticks at the center of each group
    ax.set_xticks([x + (n_columns-1) * bar_width / 2 for x in range(n_values)])
    ax.set_xticklabels(all_unique_values)
    
    # Add legend and grid
    ax.legend()
    ax.grid(axis='y', linestyle='--', alpha=0.7)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 12))

# Dynamically select columns related to tough sessions and interval sessions
tough_columns = df_week.filter(like="nr. tough sessions").columns
interval_columns = df_week.filter(like="nr. days with interval session").columns

# Create the two grouped bar charts
create_grouped_bar_chart(
    ax1, 
    tough_columns, 
    'Comparison of Tough Sessions Across Three Consecutive Weeks',
    'Number of Tough Sessions',
    'Frequency'
)

create_grouped_bar_chart(
    ax2, 
    interval_columns, 
    'Comparison of Days with Interval Sessions Across Three Consecutive Weeks',
    'Number of Days with Interval Sessions',
    'Frequency'
)

# Improve layout
plt.tight_layout()
plt.show()
```

## **6.2 Model Pipeline**

```python
# Machine Learning
# For splitting data and cross-validation
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
# For model evaluation metrics
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, accuracy_score, roc_curve, auc, roc_auc_score
# For robust scaling of numerical features
from sklearn.preprocessing import RobustScaler
# For random undersampling of the majority class
from imblearn.under_sampling import RandomUnderSampler
# For creating pipelines combining preprocessing and model training
from imblearn.pipeline import Pipeline

# Classification Models
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
```

### **6.2.1 Separating Features and Target Variable**

```python
def split_data(df, target_column, drop_columns=None, test_size=0.2, random_state=42):
    
    # Drop specified columns
    if drop_columns is None:
        drop_columns = []
        
    # Separating the features (X) and target (y)
    X = df.drop(columns=[target_column] + drop_columns)
    y = df[target_column]
    
    # Split the data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=test_size, random_state=random_state)

    return X_train, X_test, y_train, y_test
    
def scale_data(X_train, X_test):
    # Instantiate a RobustScaler
    scaler = RobustScaler()
    
    # Fit the scaler on the training data and transform the training data
    X_train_scaled = scaler.fit_transform(X_train)
    
    # Transform the test data using the fitted scaler
    X_test_scaled = scaler.transform(X_test)

    return X_train_scaled, X_test_scaled, scaler
```

```python
# Split data
X_train, X_test, y_train, y_test = split_data(df_week, target_column='injury', drop_columns=['Athlete ID'])
# Scale features
X_train_scaled, X_test_scaled, scaler = scale_data(X_train, X_test)
```

### **6.2.2 Model Selection**

```python
def evaluate_models(models, X_train, y_train, cv_splits=10, scoring_metric='recall', random_state=42):
    # List to store model names and their corresponding cross-validation scores
    names = []
    scores = []
    
    # Define StratifiedKFold to maintain class balance in each fold
    strat_k_fold = StratifiedKFold(n_splits=cv_splits, random_state=random_state, shuffle=True)
    
    # Loop over each model in the models list
    for name, model in models:
        # Create a pipeline with sampling and model
        pipeline = Pipeline([
            ('rus', RandomUnderSampler(random_state=42)),  # Undersampling
            ('model', model)  # Train model
        ])
    
        # Perform cross-validation and compute the mean recall score
        score = cross_val_score(pipeline, X_train_scaled, y_train, cv=strat_k_fold, scoring=scoring_metric).mean()
    
        # Append the model name and its score to the respective lists
        names.append(name)
        scores.append(score)
    
    # Create a DataFrame to store and display results
    cross_val_result = pd.DataFrame({'Model': names, 'Mean Recall Score': scores})
    
    # Display the cross-validation results
    return cross_val_result
```

```python
# Evaluate models
cross_val_results = evaluate_models(models, X_train_scaled, y_train)
# Display the results
cross_val_results
```

```python
def plot_recall_scores(results_df, title="Mean Recall Scores of Classifiers"):
    # Set the plot size
    plt.figure(figsize=(12, 6))
    
    # Create a barplot
    barplot = sns.barplot(x='Model', y='Mean Recall Score', data=results_df, palette='pastel', hue='Model')
    
    # Add title and labels
    plt.title(title, fontsize=14)
    plt.xlabel('Classifier', fontsize=12)
    plt.ylabel('Recall Score', fontsize=12)
    
    # Rotate x labels for better readability
    plt.xticks(rotation=45)
    
    # Annotate the bars with the accuracy score values
    for p in barplot.patches:
        barplot.annotate(format(p.get_height(), '.2f'), 
                         (p.get_x() + p.get_width() / 2., p.get_height()), 
                         ha = 'center', va = 'center', 
                         xytext = (0, 5), 
                         textcoords = 'offset points')
        
    # Display the plot
    plt.show()
```

### **6.2.4 Hyperparameter Tuning for GradientBoosting**

```python
from sklearn.model_selection import RandomizedSearchCV

def tune_gradient_boosting(X_train, y_train, n_iter=50, cv=5, random_state=42):
    
    # Define the hyperparameter grid
    param_dist = {
        'model__n_estimators': np.arange(50, 500, 50),  # Number of boosting stages
        'model__learning_rate': [0.001, 0.01, 0.05, 0.1, 0.2],  # Shrinkage factor - Smaller values (towards 0) decreases how much say each 
                                                                # weak learner has in the ensemble.
        'model__max_depth': np.arange(3, 10, 2),  # Depth of each tree
        'model__min_samples_split': np.arange(2, 10, 2),  # Minimum samples to split a node
        'model__min_samples_leaf': np.arange(1, 10, 2),  # Minimum samples per leaf - A larger minimum number of samples helps to prevent overfitting
                                                         # by making it more difficult for the trees to create splits based on too few data points.
        'model__subsample': [0.5, 0.7, 0.9, 1.0]  # Fraction of training data per boosting step - the proportion of the data used to train each tree
    }

    # Create the pipeline
    pipeline = Pipeline([
        ('rus', RandomUnderSampler(random_state=random_state)),  # Random undersampling
        ('model', GradientBoostingClassifier(random_state=random_state))  # Classifier
    ])

    # Perform Randomized Search with Pipeline
    random_search = RandomizedSearchCV(
        pipeline, 
        param_distributions=param_dist, 
        n_iter=n_iter,  
        scoring='recall',  # Optimize for recall
        cv=cv,  # Cross-validation folds
        verbose=2,
        random_state=random_state,
        n_jobs=-1  # Use all CPU cores
    )

    # Fit the search to the training data
    random_search.fit(X_train_scaled, y_train)

    # Retrieve best results
    best_pipeline = random_search.best_estimator_
    best_params = random_search.best_params_
    best_score = random_search.best_score_

    # Print the best hyperparameters and best recall score
    print("Best Hyperparameters:", random_search.best_params_)
    print("Best Recall Score:", random_search.best_score_)

    return best_pipeline
```

```python
best_pipeline = tune_gradient_boosting(X_train_scaled, y_train)
```

```python
def evaluate_model_with_cv(pipeline, X_train, y_train, scoring='recall', n_splits=10, random_state=10):
    # Define StratifiedKFold to maintain the balance of classes in each fold
    strat_k_fold = StratifiedKFold(n_splits=10, random_state=10, shuffle=True)
    
    # Perform cross-validation and compute the mean recall score
    cross_val_scores = cross_val_score(best_pipeline, X_train, y_train, cv=strat_k_fold, scoring='recall')
    
    # Print cross-validation scores
    print("Cross-validation scores:", cross_val_scores)
    print("Mean cross-validation score:", np.mean(cross_val_scores))
```
